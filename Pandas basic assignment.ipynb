**Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.**

**Pandas Series:**   
A Pandas Series is a one-dimensional array-like object that can hold data of any type (integers, floats, strings, etc.).
Each element in a Series has a unique identifier called an index.
Think of a Series as a column in a spreadsheet or a single column of a database table.
Series are fundamental building blocks for more complex structures like DataFrames.
You can create a Series from lists, arrays, dictionaries, or existing Series objects.
Example: Creating a Series from different data sources:

```python
import pandas as pd

# From a list
data_list = [1, 2, 3, 4, 5]
series_from_list = pd.Series(data_list)
print(series_from_list)

# From a dictionary
data_dict = {'a': 1, 'b': 2, 'c': 3}
series_from_dict = pd.Series(data_dict)
print(series_from_dict)

# With custom index
index = ['a', 'b', 'c', 'd', 'e']
series_custom_index = pd.Series(data_list, index=index)
print(series_custom_index)
```

**Pandas DataFrame:**

A Pandas DataFrame is a two-dimensional table-like structure consisting of multiple Series objects.
It contains rows and columns, making it suitable for tabular data.
DataFrames are commonly used for data manipulation, analysis, and exploration.
You can think of a DataFrame as a spreadsheet or a SQL table.
It can hold heterogeneous data types in different columns.

```python
# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'Salary': [50000, 60000, 45000]}
df = pd.DataFrame(data)
print(df)
```

**Q7. Create a DataFrame using multiple Series. Explain with an example.**

```python
import pandas as pd

# Create individual Series
student_names = pd.Series(["Alice", "Bob", "Charlie", "David"])
math_scores = pd.Series([85, 92, 78, 88])
english_scores = pd.Series([90, 88, 85, 92])

# Combine Series into a DataFrame
df = pd.DataFrame({
    "Name": student_names,
    "Math Score": math_scores,
    "English Score": english_scores
})

# Display the DataFrame
print(df)
```
